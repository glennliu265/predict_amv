# Plot Reanalysis Validation Results

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('corrs_for_best_models.csv')
```

```python
Ann_NAT_CNN = df['basic_CNN']
Ann_NAT_FNN = df['basic_FNN']
Ann_SPG_CNN = df['SPG_CNN']
Ann_SPG_FNN = df['SPG_FNN']
DJF_NAT_CNN = df['DJF_CNN']
DJF_NAT_FNN = df['DJF_FNN']
```

```python
leads = np.arange(0,25,1)
```

```python
fig,axs = plt.subplots(2,1,figsize=(6,8))

plt.style.use('seaborn')
plt.style.use('seaborn')
plt.style.use('seaborn-bright')
params = {'legend.fontsize': 'x-large',
         'axes.labelsize': 'x-large',
         'axes.titlesize':'x-large',
         'xtick.labelsize':'x-large',
         'ytick.labelsize':'x-large'}
plt.rcParams.update(params)

axs[0].plot(leads, Ann_NAT_CNN,linewidth=2.5,color='k')
axs[0].plot(leads, Ann_SPG_CNN,linewidth=2.5,color='cornflowerblue')
axs[0].plot(leads, DJF_NAT_CNN,linewidth=2.5,color='cyan')
axs[0].set_xlabel('Prediction Lead Time (year)')
axs[0].set_ylabel('Correlation')
axs[0].set_title('CNN2, ALL Predictors')
axs[0].set_xticks(np.arange(0,26,5))
axs[0].set_ylim([-1,1])

axs[1].plot(leads, Ann_NAT_FNN,linewidth=2.5,color='k')
axs[1].plot(leads, Ann_SPG_FNN,linewidth=2.5,color='cornflowerblue')
axs[1].plot(leads, DJF_NAT_FNN,linewidth=2.5,color='cyan')
axs[1].set_xlabel('Prediction Lead Time (year)')
axs[1].set_ylabel('Correlation')
axs[1].set_title('FNN2, ALL Predictors')
axs[1].set_xticks(np.arange(0,26,5))
axs[1].set_ylim([-1,1])


axs[1].legend(['Annual North Atlantic (NAT)','Annual Subpolar Gyre (SPG)','Winter (DJF) North Atlantic (NAT)'])
plt.suptitle("Applying Sub-region, Season NNs on Reanalysis Data",fontsize=16)
plt.tight_layout()



#plt.savefig('../Figures/Fig5_corr_on_reanalysis.png',dpi=300)
```
