```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('corrs_for_best_models.csv')
```

```python
Ann_NAT_CNN = df['basic_CNN']
Ann_NAT_FNN = df['basic_FNN']
Ann_SPG_CNN = df['SPG_CNN']
Ann_SPG_FNN = df['SPG_FNN']
DJF_NAT_CNN = df['DJF_CNN']
DJF_NAT_FNN = df['DJF_FNN']
```

```python
leads = np.arange(0,25,1)
```

```python
fig,axs = plt.subplots(1,2,sharey=True,figsize=(10,5))
plt.style.use('seaborn')
params = {'legend.fontsize': 'small',
         'axes.labelsize': 'small',
         'axes.titlesize':'small',
         'xtick.labelsize':'small',
         'ytick.labelsize':'small'}
plt.rcParams.update(params)

axs[0].plot(leads, Ann_NAT_CNN,linewidth=2)
axs[0].plot(leads, Ann_SPG_CNN,linewidth=2)
axs[0].plot(leads, DJF_NAT_CNN,linewidth=2)
axs[0].set_xlabel('Lead (yr)')
axs[0].set_ylabel('Correlation')
axs[0].set_title('CNN2')
axs[0].set_xticks(np.arange(0,26,5))

axs[1].plot(leads, Ann_NAT_FNN,linewidth=2)
axs[1].plot(leads, Ann_SPG_FNN,linewidth=2)
axs[1].plot(leads, DJF_NAT_FNN,linewidth=2)
axs[1].set_xlabel('Lead (yr)')
axs[1].set_title('FNN2')
axs[1].set_xticks(np.arange(0,26,5))

axs[1].legend(['Ann_NAT','Ann_SPG','DJF_NAT'])
plt.tight_layout()
plt.savefig('Fig5_corr_on_reanalysis.png',dpi=300)
```
