```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
path = '../../CESM_data/resnet50_results/'
```

```python
leads = np.arange(0,25,1)

fnames = os.listdir(path)
fnames.sort()
fnames
```

```python
corr_grid_test = []
corr_grid_train = []

for fname in fnames:
    f = np.load(path+fname)
    corr_grid_test.append( f['test_corr'] )
    corr_grid_train.append( f['train_corr'] )
```

```python
plt.plot(leads, corr_grid_train)
plt.plot(leads, corr_grid_test)
plt.title('ResNet50 (20 ensembles, 10 epochs) correlation')
plt.legend(['train corr','test corr'])
plt.xlabel('lag (year)')
plt.ylabel('correlation')
```
