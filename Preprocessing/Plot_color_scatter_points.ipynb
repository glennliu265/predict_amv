```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
```

```python
sst_normed = np.load('../../CESM_data/CESM_SST_normalized_lat_weighted.npy').astype(np.float32)
sss_normed = np.load('../../CESM_data/CESM_SSS_normalized_lat_weighted.npy').astype(np.float32)
```

```python
def get_data_at_lead(lead):
    y = np.mean(sst_normed[:,lead:,:,:],axis=(2,3)).reshape((tstep-lead)*42,1)  
    X = np.transpose(
        np.array([sst_normed,sss_normed])[:,:,0:tstep-lead,:,:].reshape(2,(tstep-lead)*42,33,89),
        (1,0,2,3))
    return X,y
```

```python
tstep = 1032
```

```python
plt.figure(figsize=(5,4))
plt.rcParams.update({'font.size': 12})

lead = 0
sc = plt.scatter(np.mean(get_data_at_lead(lead)[0][:,0,:,:],axis=(1,2)),
                 np.mean(get_data_at_lead(lead)[0][:,1,:,:],axis=(1,2)),
                 c=get_data_at_lead(lead)[1],vmin=-1,vmax=1,cmap='RdBu_r',s=20,alpha=0.8)

plt.colorbar(sc,label='Future SST (std)')
plt.ylim([-1.2,1.2])
plt.xlim([-1.2,1.2])
plt.xlabel('Lagged SST (std)')
plt.ylabel('Lagged SSS (std)')
plt.title('Future SST '+str(lead)+' months lead')
plt.tight_layout()
#plt.savefig('../Figures/colored_scatter_data_lead'+str(lead)+'.pdf')
plt.show()
```
